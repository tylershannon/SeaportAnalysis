## Search for Restaurants

```python
#define location inputs
google_key = "AIzaSyBc-u0CIErx03OgmSzA3jRsLVWnGQzGPnU"

lat = "42.346939"
long = "-71.044458"
radius = "1000"
key = google_key
amenity_type = "restaurant"

search_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius={}&key={}&type={}".format(lat, long, radius, key, amenity_type)
```

### page 1

```python
urlData = search_url
webURL = urllib.request.urlopen(urlData)
data = webURL.read()
encoding = webURL.info().get_content_charset('utf-8')
response = json.loads(data.decode(encoding))
```

```python
print(response)
```

```python
venues = {}
counter = 0

for item in response['results']:
    venue = {
        'name' : item['name'],
        'type' : item['types'],
        'lat' : item['geometry']['location']['lat'],
        'long' : item['geometry']['location']['lng'],
        'id' : item['place_id']}
    venues['item_{}'.format(counter)] = venue
    counter += 1
```

### page 2

```python
urlData = search_url + "&pagetoken=" + response["next_page_token"]
webURL = urllib.request.urlopen(urlData)
data = webURL.read()
encoding = webURL.info().get_content_charset('utf-8')
response2 = json.loads(data.decode(encoding))
```

```python
print(response2)
```

```python
counter = 20
for item in response2['results']:
    venue = {
        'name' : item['name'],
        'type' : item['types'],
        'lat' : item['geometry']['location']['lat'],
        'long' : item['geometry']['location']['lng'],
        'id' : item['place_id']}
    venues['item_{}'.format(counter)] = venue
    counter += 1
```

### page 3

```python
urlData = search_url + "&pagetoken=" + response2["next_page_token"]
webURL = urllib.request.urlopen(urlData)
data = webURL.read()
encoding = webURL.info().get_content_charset('utf-8')
response3 = json.loads(data.decode(encoding))
```

### gather popular time data for each location

```python
import populartimes
keys = []
for key in venues.keys():
    keys.append(key)
```

```python
venues_pop_times = []
for i in range(len(venues)):
    id = venues[keys[i]]['id']
    a = populartimes.get_id(google_key, id)
    venues_pop_times.append(a)
    print(((i+1)/(len(venues)))*100, "% complete")
```

```python
venues_pop_times
```

```python
venues_pop_times[0]
```

```python
venues_final = []
for i in range(len(venues_pop_times)):
    try:
        if venues_pop_times[i]['populartimes']:
            venues_final.append(venues_pop_times[i])
    except KeyError:
        pass
venues_final
```

```python
len(venues_final)
```

```python
with open('data.json', 'w') as outfile:  
    json.dump(venues_final, outfile, sort_keys=True, indent=4)
```

```python
Monday = []
Tuesday = []
Wednesday = []
Thursday = []
Friday = []
Saturday = []
Sunday = []

for i in range(len(venues_final)):
    sum_mon = sum(venues_final[i]['populartimes'][0]['data'])
    sum_tue = sum(venues_final[i]['populartimes'][1]['data'])
    sum_wed = sum(venues_final[i]['populartimes'][2]['data'])
    sum_thu = sum(venues_final[i]['populartimes'][3]['data'])
    sum_fri = sum(venues_final[i]['populartimes'][4]['data'])
    sum_sat = sum(venues_final[i]['populartimes'][5]['data'])
    sum_sun = sum(venues_final[i]['populartimes'][6]['data'])
    Monday.append(sum_mon)
    Tuesday.append(sum_tue)
    Wednesday.append(sum_wed)
    Thursday.append(sum_thu)
    Friday.append(sum_fri)
    Saturday.append(sum_sat)
    Sunday.append(sum_sun)
```

```python
counts = {
    'Day': ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday',],
    'Counts': [sum(Monday), sum(Tuesday),sum(Wednesday),sum(Thursday),sum(Friday),sum(Saturday),sum(Sunday)]
}
df = pd.DataFrame(counts)
df.head()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="Day", y="Counts", data=df, color='#000000')
plt.savefig('businessActivityAggregate.png')
```

```python
venue_hour_counts = []
for h in range(24):
    hours_count = 0
    for d in range(7):
        for a in range(len(venues_final)):
            hour = venues_final[a]['populartimes'][d]['data'][h]
            hours_count = hours_count+hour
    venue_hour_counts.append(hours_count)

venue_hour_counts
```

```python
hour_counts = {
    'Hour': ['0:00','1:00','2:00','3:00','4:00','5:00','6:00','7:00','8:00','9:00','10:00','11:00','12:00','13:00'
            ,'14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00'],
    'Counts': venue_hour_counts
}
df = pd.DataFrame(hour_counts)
df.head()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
ax = sns.barplot(x="Hour", y="Counts", data=df, color='#000000')
plt.xticks(rotation=45)
plt.savefig('businessActivityHourly.png')
```

```python
Monday = []
Tuesday = []
Wednesday = []
Thursday = []
Friday = []
Saturday = []
Sunday = []

for i in range(len(convention_pop_times)):
    sum_mon = sum(convention_pop_times[i]['populartimes'][0]['data'])
    sum_tue = sum(convention_pop_times[i]['populartimes'][1]['data'])
    sum_wed = sum(convention_pop_times[i]['populartimes'][2]['data'])
    sum_thu = sum(convention_pop_times[i]['populartimes'][3]['data'])
    sum_fri = sum(convention_pop_times[i]['populartimes'][4]['data'])
    sum_sat = sum(convention_pop_times[i]['populartimes'][5]['data'])
    sum_sun = sum(convention_pop_times[i]['populartimes'][6]['data'])
    Monday.append(sum_mon)
    Tuesday.append(sum_tue)
    Wednesday.append(sum_wed)
    Thursday.append(sum_thu)
    Friday.append(sum_fri)
    Saturday.append(sum_sat)
    Sunday.append(sum_sun)
    
    
counts = {
    'Day': ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday',],
    'Counts': [sum(Monday), sum(Tuesday),sum(Wednesday),sum(Thursday),sum(Friday),sum(Saturday),sum(Sunday)]
}
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
df1 = pd.DataFrame(counts)
#df.head()

sns.set_style("whitegrid")
ax = sns.barplot(x="Day", y="Counts", data=df1, color='#000000')
plt.savefig('ConventionActivityAggregate.png')
```

```python
convention_hour_counts = []
for h in range(24):
    hours_count = 0
    for d in range(7):
        for a in range(len(convention_pop_times)):
            hour = convention_pop_times[a]['populartimes'][d]['data'][h]
            hours_count = hours_count+hour
    convention_hour_counts.append(hours_count)

hour_counts = {
    'Hour': ['0:00','1:00','2:00','3:00','4:00','5:00','6:00','7:00','8:00','9:00','10:00','11:00','12:00','13:00'
            ,'14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00'],
    'Counts': convention_hour_counts
}
df2 = pd.DataFrame(hour_counts)


import seaborn as sns
sns.set_style("whitegrid")
ax = sns.barplot(x="Hour", y="Counts", data=df2, color='#000000')
plt.xticks(rotation=45)
plt.savefig('ConventionActivityHourly.png')
```

```python
days = []
hours = []
hour_count = []
for h in range(24):
    for d in range(7):
        hour_count1 = convention_pop_times[0]['populartimes'][d]['data'][h]
        hour = h
        day = d
        days.append(day)
        hours.append(h)
        hour_count.append(hour_count1)
```

```python
data = {
    'Days': days,
    'Hours': hours,
    'Counts': hour_count
}
df3 = pd.DataFrame(data)

        
        
df3.head()
```
